TASK 2-Build Preprocessing Function

```python
def preprocess_text(text):
    if not text or not isinstance(text, str):
        return [], ""

    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()
    tokens = [w for w in tokens if len(w) > 2 or w in ['no', 'not']]

    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
```

TASK 3 - Stress TESTING

```python
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for s in sentences:
    tokens, cleaned = preprocess_text(s)

    print("Original:", s)
    print("Tokens:", tokens)
    print("Cleaned:", cleaned)
    print("-"*50)

    results.append((s, tokens, cleaned))
```

TASK 4- Token Analytics

```python
for original, tokens, cleaned in results:
    avg_len = sum(len(w) for w in tokens)/len(tokens) if tokens else 0

    print("Sentence:", original)
    print("Total Tokens:", len(tokens))
    print("Unique Tokens:", len(set(tokens)))
    print("Avg Length:", round(avg_len,2))
    print("-"*50)
```

## Analysis

Most noisy sentence:

Most meaningful sentence:

TASK 5 - Frequency Analysis

```python
all_tokens = []

for _, tokens, _ in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 Frequent Words:")
print(counter.most_common(10))

print("\nLeast 5 Words:")
print(counter.most_common()[-5:])
```

TASK 6 - Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
output = full_pipeline(sentences)
print(output)
```
